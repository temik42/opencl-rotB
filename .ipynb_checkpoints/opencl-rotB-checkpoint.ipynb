```python
import pyopencl as cl
from pyopencl import array
import numpy as np



class CL:
    def __init__(self, X, B):        
        self.shape = X.shape[0:3]
        self.size = X.nbytes      
        self.ctx = cl.create_some_context()
        self.queue = cl.CommandQueue(self.ctx)
        self.mf = cl.mem_flags
        self.fstr = "".join(open('rotB.cl', 'r').readlines())
        self.program = cl.Program(self.ctx, self.fstr).build()
        self.X = cl.Buffer(self.ctx, self.mf.READ_ONLY | self.mf.COPY_HOST_PTR, hostbuf=X)
        self.B = cl.Buffer(self.ctx, self.mf.READ_ONLY | self.mf.COPY_HOST_PTR, hostbuf=B)
        #self.temp = cl.Buffer(self.ctx, self.mf.READ_WRITE | self.mf.COPY_HOST_PTR, hostbuf = np.zeros_like(X))  

    def rotB(self):
        self.temp = cl.Buffer(self.ctx, self.mf.WRITE_ONLY, self.size)  
        out_buf = cl.Buffer(self.ctx, self.mf.WRITE_ONLY, self.size)   
        self.program.rotB(self.queue, self.shape, None, self.X, self.B, self.temp, out_buf)
        return out_buf    
        
    #def d(self, in_buf, dim, m):
    #    out_buf = cl.Buffer(self.ctx, self.mf.WRITE_ONLY, self.size)        
    #    self.program.deriv(self.queue, self.shape, None, in_buf, np.byte(dim), np.byte(m), out_buf)
    #    return out_buf
    
    #def Jacobian(self):
    #    J = [self.d(self.X, i, 1) for i in range(0,3)]
    #    return J
            
    #def Hessian(self):
    #    J = self.Jacobian()
    #    H = [[self.d(self.X, j, 2) if (j == k) else
    #          self.d(J[j], k, 1) for k in range(0,3)] for j in range(0,3)]
    #    return (H, J)
        



if __name__ == "__main__":

    dim = (64,64,64)
    idx = np.indices(dim).astype(np.float32)  
    X = np.zeros(dim+(4,), dtype = np.float32)
    X[:,:,:,0] = idx[0,:,:,:]
    X[:,:,:,1] = idx[1,:,:,:]
    X[:,:,:,2] = idx[2,:,:,:]
    
    X1 = np.array(X)
    ar = 16**2
    az = 16
    h = np.exp(-((X[:,:,:,0]-dim[0]/2)**2 + (X[:,:,:,1]-dim[1]/2)**2)/ar - X[:,:,:,2]/az)*np.pi/4

    X[:,:,:,0] = (X1[:,:,:,0]-dim[0]/2)*np.cos(h) - (X1[:,:,:,1]-dim[1]/2)*np.sin(h) + dim[0]/2
    X[:,:,:,1] = (X1[:,:,:,0]-dim[0]/2)*np.sin(h) + (X1[:,:,:,1]-dim[1]/2)*np.cos(h) + dim[1]/2
    
    B = np.zeros(dim+(4,), dtype = np.float32)
    B[:,:,:,2] = np.ones(dim, dtype = np.float32)
    #B = [np.zeros(dim) for i in range(0,3)]
    #B[2] = np.ones(dim)
    
    #a = np.random.random(dim + (4,))#.astype(np.float32)
    #a = np.array(X).astype(np.float32)
    
    _CL = CL(X, B)
    
    c = np.zeros(dim + (4,), dtype = np.float32)
    cl.enqueue_read_buffer(_CL.queue, _CL.rotB(), c).wait()
    
    
    #H = _CL.Hessian()[1]
    
    #for i in range(0,3):
    #    for j in range(0,3):
    #        c = np.empty_like(a[0])
    #        cl.enqueue_read_buffer(_CL.queue, H[0][i][j], c).wait()
        
    #        print c
```
